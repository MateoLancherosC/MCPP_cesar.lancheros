## DISTRIBUCIÓN DE FRECUENCIA

```python
#Importar Distribución de Frecuencia

from nltk.probability import FreqDist
```

```python
#Definir frecuencia PND 2010-2014

FrecuenciaPND1 = FreqDist(PND_1T)
```

```python
#Definir frecuencia PND 2014-2018

FrecuenciaPND2 = FreqDist(PND_2T)
```

```python
#Definir frecuencia PND 2018-2022

FrecuenciaPND3 = FreqDist(PND_3T)
```

```python
#Plot de frecuencia PND 2010-2014

FrecuenciaPND1.plot(20,cumulative=False);
```

```python
#Guardar imagen de frecuancia PND 2010-2014

plt.savefig("FrecuenciaPND1.png");
```

```python
#Plot de frecuencia PND 2014-2018

FrecuenciaPND2.plot(20,cumulative=False)

plt.savefig;("FrecuenciaPND2.png");
```

```python
#Guardar imagen de frecuancia PND 2014-2018

plt.savefig;("FrecuenciaPND2.png");
```

```python
#Plot de frecuencia PND 2018-2022

FrecuenciaPND3.plot(20,cumulative=False)
```

```python
#Guardar imagen de frecuancia PND 2018-2022

plt.savefig("FrecuenciaPND3.png");
```

```python
PND = PND_1T + PND_2T + PND_3T
```

```python
# Installar libreria para nube de palabras

from os import path
from wordcloud import WordCloud
```

```python
#Nube de palabras PND 2010-2014

d = path.dirname('‎⁨iCloud Drive⁩ ▸ ⁨Documentos⁩ ▸ ⁨Universidad del Rosario⁩ ▸ ⁨Maestría en Economía de las Políticas Públicas⁩ ▸ ⁨Métodos Computacionales⁩ ▸ ⁨GitHub⁩ ▸ ⁨MCPP_cesar.lancheros⁩ ▸ ⁨Proyecto Final⁩ ▸ ⁨Código⁩')
# Read the whole text
text = open(path.join(d, 'PND_1str.txt')).read()

wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")

# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
plt.savefig('WordCloudPND1.png');
```
