## LECTURA DE TEXTOS

```python
#Importar docx para procesar los PND en formato word

import docx
```

```python
def readtxt(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
#Leer PND 2010-2014. Tomo I

PND1 = readtxt('PND_2010_2014_I.docx')
```

```python
#Leer PND 2010-2014. Tomo II

PND11 = readtxt('PND_2010_2014_II.docx')
```

## LIMPIEZA DE LOS TEXTOS

```python
#Definir puntuación

string.punctuation
puntuacion= string.punctuation + "–¡¿”“•\r"
puntuacion
```

```python
#Definir eliminación de puntuación

def no_punct(string):
    sinpuntuacion = string.maketrans('', '', puntuacion)
    return string.translate(sinpuntuacion)
```

```python
# Quitar puntuacion PND 2010-2014

PND_1 = no_punct(PND_1)
```

```python
PND_1[0:2000]
```

```python
#Minisculas PND 2010-2014

PND_1 = PND_1.lower()
```

```python
#Importar código para tildes

import unidecode
```

```python
#Reemplazar tildes PND 2010-2014

PND_1 = unidecode.unidecode(PND_1)
```

```python
# Definir función para limpiar números y espaciado usando expresiones regulares

def clean_PND(text):
    return re.sub('(\\n)|([0-9])',' ',text)
```

```python
#Limpiar números y espaciado PND 2010-2014

PND_1 = clean_PND(PND_1)
```
